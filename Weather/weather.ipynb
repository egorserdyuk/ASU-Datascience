```python
import pandas as pd
```

```python
df = pd.read_excel("data.xlsx", names=['Индекс ВМО', 'Год', 'Месяц', 'День', 'Минимиальная температура воздуха', 'Средняя температура воздуха', 'Максимальная температура воздуха', 'Количество осадков'])
```

```python
df.head()
```

```python
df['Дата'] = df['Год'].astype(str) + df['Месяц'].astype(str) + df['День'].astype(str)
df = df.drop('Год', axis=1)
df = df.drop('Месяц', axis=1)
df = df.drop('День', axis=1)

df['Дата'] = pd.to_datetime(df['Дата'], format='%Y%m%d')
```

```python
df = df.set_index(df['Дата'])

df.head()
```

```python
df = df.drop('Индекс ВМО', axis=1)

df.head()
```

```python
print('Date lost =' + str(pd.isnull(df['Дата']).sum()))
print('Min lost =' + str(pd.isnull(df['Минимиальная температура воздуха']).sum()))
print('Mean lost =' + str(pd.isnull(df['Средняя температура воздуха']).sum()))
print('Max lost =' + str(pd.isnull(df['Максимальная температура воздуха']).sum()))
print('Value lost =' + str(pd.isnull(df['Количество осадков']).sum()))
```

```python
for year in range(1959, 2017):
    if df[df.index.year == year].isnull().sum().sum() > 0:
        print(year, df[df.index.year == year].isnull().sum().sum())
```
