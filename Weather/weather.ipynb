```python
import pandas as pd
```

```python
df = pd.read_excel("data.xlsx", names=['Индекс ВМО', 'Год', 'Месяц', 'День', 'Минимиальная температура воздуха', 'Средняя температура воздуха', 'Максимальная температура воздуха', 'Количество осадков'])
```

```python
df.head()
```

```python
df['Дата'] = df['Год'].astype(str) + df['Месяц'].astype(str) + df['День'].astype(str)

df['Дата'] = pd.to_datetime(df['Дата'], format='%Y%m%d')
```

```python
df = df.set_index(df['Дата'])

df.head()
```

```python
df = df.drop('Индекс ВМО', axis=1)

df.head()
```

```python
print('Date lost =' + str(pd.isnull(df['Дата']).sum()))
print('Min lost =' + str(pd.isnull(df['Минимиальная температура воздуха']).sum()))
print('Mean lost =' + str(pd.isnull(df['Средняя температура воздуха']).sum()))
print('Max lost =' + str(pd.isnull(df['Максимальная температура воздуха']).sum()))
print('Value lost =' + str(pd.isnull(df['Количество осадков']).sum()))
```

```python
for year in range(1959, 2017):
    if df[df.index.year == year].isnull().sum().sum() > 0:
        print(year, df[df.index.year == year].isnull().sum().sum())
```

```python
df
```

```python
df = df.sort_index()
df = df.truncate(before=pd.Timestamp('1966-01-01'))
df
```

```python
df['Размах температур'] = df['Максимальная температура воздуха'] - df['Минимиальная температура воздуха']
df
```

```python
df["Количество дней без осадков"] = 0
x = 0
for index, row in df.iterrows():
    if row["Количество осадков"] == 0:
        row["Количество дней без осадков"] += x
        x += 1
    else:
        x = 0
    df.loc[index, "Количество дней без осадков"] = row["Количество дней без осадков"]
    
df
```

```python
maxPer = df[df["Количество дней без осадков"] == df["Количество дней без осадков"].max()][
    ["Дата", "Количество дней без осадков"]]

maxPer
```

```python
avrJune = df[df["Месяц"] == 6].groupby(["Год", "Месяц"])["Средняя температура воздуха"].mean()

avrJune
```

```python
maxJan = df[df["Месяц"] == 1].groupby(["Год", "Месяц"])["Максимальная температура воздуха"].max()

maxJan
```

```python
df = df.drop('Год', axis=1)
df = df.drop('Месяц', axis=1)
df = df.drop('День', axis=1)
```

```python
p1 = df[df["Средняя температура воздуха"] < -30]
p2 = df[(df["Средняя температура воздуха"] > 25) & (df["Количество дней без осадков"] > 3)]
```
