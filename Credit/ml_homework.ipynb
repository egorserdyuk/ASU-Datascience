```python
import numpy as np
import pandas as pd

from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split as tts
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_excel("Credit.xlsx")
df
```

```python
df = df.replace({"Да": 1, "Нет": 0})  # Бинарные на числовые
df = df.drop(['Количество', 'Дата кредитования'], axis=1)  # Значение параметра всегда одно и то же, дата не влияет
df
```

```python
fact = {}
for col in [
            "Пол",
            "Расположение",
            "Должность",
            "Специализация",
           ]:
    df[col], fact[col] = pd.factorize(df[col])
```

```python
df["Образование"] = df["Образование"].map({"среднее":0, "специальное": 1, "высшее":2})
df["Машина"] = df["Машина"].map({"Нет автомобиля":0, "отечественная": 1, "импортная":2})
df["Класс предприятия"] = df["Класс предприятия"].map({"малое":0, "среднее": 1, "крупное":2})
```

```python
labels = LabelEncoder()
for col in ["Цель кредитования", "Способ приобретения собств.", "Отрасль предприятия", "Основное направление расходов"]:
    labels.fit(df[col])
    df[col] = labels.transform(df[col])
df
```

```python
df  # Все переменные стали числовые
```

```python
y = df["Давать кредит"]
x = df.drop("Давать кредит", axis=1)
x_train, x_test, y_train, y_test = tts(x, y, test_size=0.3, random_state=40)  # Разделяем выборку
```

```python
tree_clf = DecisionTreeClassifier()
tree_clf.fit(x_train, y_train)
```

```python
params = {
    "criterion" : ["gini", "entropy"],
    "max_depth" : [None, 1, 2, 3, 10]
}

# Поиск наилучших параметров делаем по точности(precision, полноте(recall) среднему гармоническому 
# precision и recall(f1), последний из которых является результирующим в подборе параметров
# Precision и recall не зависят, в отличие от accuracy, от соотношения классов и потому применимы в условиях несбалансированных выборок
# А f1 как гармноческое соотношение между ними, очевидно, хороший параметр, который следует улучшать
```

```python
GridSCV = GridSearchCV(tree_clf, params, scoring= ['f1', 'precision', 'recall'], refit="f1", cv=10)
GridSCV.fit(x, y)
```

```python
GridSCV.best_score_
```

```python
f_imp = pd.DataFrame({"Признак": x.columns, "Информативность": GridSCV.best_estimator_.feature_importances_})
f_imp = f_imp[f_imp["Информативность"]>0.00].sort_values("Информативность", ascending=False)
f_imp
```

```python
df_n = df.drop(f_imp['Признак'][f_imp['Информативность'] < 0.05], axis=1)  # Выбрасывем шумовые признаки
```

```python
y = df_n["Давать кредит"]
x = df_n.drop("Давать кредит", axis=1)
x_train, x_test, y_train, y_test = tts(x, y, test_size=0.3, random_state=40)  # Разделяем выборку

tree_clf = DecisionTreeClassifier()
tree_clf.fit(x_train, y_train)  # Обучаем модель

GridSCV_n = GridSearchCV(tree_clf, params, scoring= ['f1', 'precision', 'recall'], refit="f1", cv=10)
GridSCV_n.fit(x, y)
```

```python
GridSCV_n.best_score_ - GridSCV.best_score_  
# Качество классификации улучшилось после исключения шумовых признаков, но не значительно
```

Методы улучшения качества классификации дают прирост, но небольшой. В совокупности, при правильном применении этих методов, улучшение качества классификации может дать значительный результат
